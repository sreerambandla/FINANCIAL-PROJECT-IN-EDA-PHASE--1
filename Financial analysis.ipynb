This sunburst chart visualizes the average profit distribution of various products within segments, segmented by country. The multi-layered design reveals insights into how each countrys segments and specific products contribute to overall profit. Heres how this chart is helpful and some insights it provides:

### How to Use the Chart:
1. **Overall Profit Comparison by Country**: Each major slice of the sunburst represents a different country, enabling a quick view of which countries contribute the most or least to total profit.
2. **Segment-Level Profit Insight**: The second layer shows different segments (e.g., Government, Small Business) within each country, helping identify which segment drives the most profit in each country.
3. **Product-Level Profit Analysis**: The innermost layer shows the distribution of profits among products within each segment and country, allowing for a detailed breakdown of which products are the highest or lowest profit contributors.

### Insights from the Chart:

1. **High Profit from Government Segments**:

### Usage of the Chart:
This chart is useful for stakeholders to:
   - Identify high-profit segments and products per country.
   - Make informed decisions on regional market strategies and investments.
   - Prioritize resource allocation by focusing on high-performing products and segments within specific countries.
   - Spot trends in product performance that vary by segment and adjust marketing or sales strategies accordingly.

By visualizing profit distribution in this detailed, hierarchical manner, decision-makers can quickly grasp complex data relationships and make strategic adjustments at both country and product levels.

```python
def country_product_profit(country,segment,product):
  data = pd.DataFrame(df2014.groupby(['Country','Segment','Month Name','Product'])['Profit'].mean().reset_index())
  con = data[data['Country'] == country ]
  seg = con[con['Segment'] == segment ]
  final_data = seg[seg['Product'] == product ]
  final_data.columns = ['Country','Segment','Month','Product', 'average profit']
  final_data['Month'] = pd.Categorical(final_data['Month'], categories = month_order, ordered=True)
  final_data = final_data.sort_values('Month')
  fig = px.bar(final_data, x='Month', y='average profit', text_auto=True,
                title= f"{product} product of {segment} Segment average profit for each month of 2014 in {country}")
  return fig.show()
```

```python
country_product_profit('Germany','Government','Paseo')
```
